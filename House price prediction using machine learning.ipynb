```python
# handle categorical values using label encoder

# Import label encoder
from sklearn import preprocessing

# Create a label encoder object
label_encoder = preprocessing.LabelEncoder()

# Encode label in the mainroad column
# df["mainroad_sample"] = label_encoder.fit_transform(df["mainroad"])

categorical_cols = df.select_dtypes(include=['object']).columns

for x in categorical_cols :
    df[x] = label_encoder.fit_transform(df[x])

df
```

```python
# train test split
from sklearn.model_selection import train_test_split

# spliting the data into dependant and independant variable 
X = df.drop(columns = ["price"])
y = df["price"]

# setting the train test variables 
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.2 , random_state = 10)
```

```python
# train machine learning model linera regression
model = linear_model.LinearRegression()

# train the linear regression model
model.fit(X_train , y_train)

# reg.score will show how much accurate our model is 0.7331102153370128 = 73 percent 
model.score(X_test , y_test)
```
